```python
from audiocraft.models import MusicGen
from audiocraft.data.audio import audio_write
```

GET 5 to 10 or more prompts paste below and run two cells

```python
prompts = [
    "Sultry bossa nova groove with smooth vocals and rhythmic guitar strumming.",
    "High-energy pop-punk anthem with catchy hooks and driving power chords.",
    "Mysterious trip-hop track with haunting vocals and atmospheric beats.",
    "Frenetic drum and bass banger with intense rhythms and blistering basslines.",
    "Quirky electro-swing number with swinging beats and playful horn arrangements.",
    "Epic fantasy soundtrack with sweeping orchestration and heroic themes.",
    "Soothing lo-fi jazz instrumental with laid-back saxophone and mellow piano.",
    "Futuristic synth-pop ballad with dreamy synths and emotive vocals.",
    "Vibrant Bollywood dance number with energetic percussion and infectious melodies.",
    "Surreal ambient sound collage with layered textures and ethereal voices."
]
```

```python
model = MusicGen.get_pretrained('facebook/musicgen-medium')
model.set_generation_params(duration=10) 
wav = model.generate(prompts)
```

```python
for idx, one_wav in enumerate(wav):
    # Will save under {idx}.wav, with loudness normalization at -14 db LUFS.
    # Add idx + the starting index of the prompts given in generate line
    audio_write(f'/kaggle/working/TDL-11/{prompts[idx]}', one_wav.cpu(), model.sample_rate, strategy="loudness", loudness_compressor=True)
```
